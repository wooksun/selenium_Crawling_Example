```python
import warnings
warnings.filterwarnings(action='ignore')
%config Computer.use_hedi = False
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
```

```python
option = Options()
option.add_argument('--window-size=1890,1030') # 가상 크롬의 창 크기를 지정한다.
driver = webdriver.Chrome('./chromedriver.exe', options=option) # 가상 크롬을 실행한다.
driver.implicitly_wait(5)

# 네이버 웹툰 신혼일기 22화 다이어트 편을 보여주는 페이지
# https://comic.naver.com/webtoon/detail?titleId=812354&no=22
driver.get('https://comic.naver.com/webtoon/detail?titleId=812354&no=22')
html = driver.page_source # 가상 크롬에 실행된 사이트의 html 정보를 얻어온다.
soup = BeautifulSoup(html, 'html.parser') # 읽은 html 문서를 파싱할 수 있도록 준비한다.
reviews = soup.select('span.u_cbox_contents')
print(len(reviews))

# 베댓의 개수 만큼 반복하며 베댓과 베댓의 답글을 읽는다.
for i in range(len(reviews))[:3]:
    print('{:2d}번째 베댓 {}'.format(i + 1, reviews[i].text))
    print('=' * 80)
    
    driver.find_element_by_xpath(
        '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[' + str(i + 1) + ']/div[1]/div/div[4]/a'
    ).click()
    time.sleep(0.5)
    print('=' * 80)
    
    while True:
        try:
            driver.find_element_by_xpath(
                '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[{}]/div[2]/div[2]/a'.format(i + 1)
            ).click()
            time.sleep(0.5)
        except:
            break
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    replys = soup.select('#cbox_module_wai_u_cbox_content_wrap_tabpanel div.u_cbox_reply_area span.u_cbox_contents')
    print(len(replys))
    for reply in replys:
        print(reply.text.split('\n'))
    print('{:2d}번째 베댓의 답글'.format(i + 1))
    print('=' * 80)
    
    driver.find_element_by_xpath(
        '//*[@id="cbox_module_wai_u_cbox_content_wrap_tabpanel"]/ul/li[' + str(i + 1) + ']/div[2]/a' # 댓글 접기
    ).click()
    time.sleep(0.5)
```
